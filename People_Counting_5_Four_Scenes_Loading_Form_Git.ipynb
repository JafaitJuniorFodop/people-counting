```python
import pandas as pd
import numpy as np

import random

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#function to make scatter plots of the measurements
def plot_file(df, indices, ax):
    idx = random.choice(indices)
    files = df.iloc[idx].Files_numbers.unique()

    file = np.random.choice(files, replace=False)

    X = df.iloc[idx].iloc[np.where(df.iloc[idx].Files_numbers==file)[0]]
    #print(X.Labels.unique())
    columns = np.random.choice(np.array(['X', 'Y', 'Z']), size=2, replace=False)

    

    sns.scatterplot(ax=ax, data=X, x=columns[0], y=columns[1])
    ax.set_title('Counts: ' + str(X.Labels.iloc[0]))
```

```python
url = 'https://raw.githubusercontent.com/JafaitJuniorFodop/people-counting/main/meas_radar.csv'

df = pd.read_csv(url)
```

```python
#indices corresponding to the files with 0, 1, and 2 peoples respectively
idx_0 = np.where(df.Labels.values==0)[0]
idx_1 = np.where(df.Labels.values==1)[0]
idx_2 = np.where(df.Labels.values==2)[0]
```

```python
#2D plot. The axis X, Y and Z are choosen randomly. We are considering to plot in 3D
indices = [idx_0, idx_1, idx_2]

_, axes = plt.subplots(2, 2, figsize=(16,12))
for row in axes:
    for ax in row:
        plot_file(df, indices, ax)
```
